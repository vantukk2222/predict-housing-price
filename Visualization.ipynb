```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

data = pd.read_csv('data.csv')
```

Biểu đồ phân bố các đặc trưng trước khi xử lý ngoại lệ

```python
fig = plt.figure(figsize=(16, 16))
gs = gridspec.GridSpec(3, 3)

# Đồ thị thứ nhất
ax1 = fig.add_subplot(gs[0, 0])
sns.histplot(data=data['area'], ax=ax1, kde=True)

# Đồ thị thứ hai
ax2 = fig.add_subplot(gs[0, 1])
sns.histplot(data=data['room'], ax=ax2, kde=False)

# Đồ thị thứ ba
ax3 = fig.add_subplot(gs[0, 2])
sns.histplot(data=data['toilet'], ax=ax3, kde=False)

# Đồ thị thứ tư
ax4 = fig.add_subplot(gs[1, 0])
sns.histplot(data=data['loai_bds-code'], ax=ax4, kde=False)

# Đồ thị thứ sáu
ax6 = fig.add_subplot(gs[1, 2])
sns.histplot(data=data['price-VND'], ax=ax6, kde=True)

# Đồ thị thứ bảy
ax7 = fig.add_subplot(gs[2, 0])
sns.histplot(data=data['quan-code'], ax=ax7, kde=False)
```
